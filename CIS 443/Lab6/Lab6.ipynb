Create a linear regression model using __linregress__ from SciPy. Use the model to predict the high temperature in August 2025.

```python
# Put Lab 6 code here
import pandas as pd
from scipy import stats
import seaborn as sns

louisville_tem = pd.read_csv('LouisvilleTem-1948-2020.csv')
year = 2025

louisville_tem.columns = ['Date', 'Temperature', 'Anomaly']
louisville_tem.Date = louisville_tem.Date.floordiv(100)
linear_regression = stats.linregress(x=louisville_tem.Date, y=louisville_tem.Temperature) 
slope = linear_regression.slope
intercept = linear_regression.intercept
predict_2025 = linear_regression.slope * year + linear_regression.intercept

print(f' Predicted highest temperature in August 2025: {predict_2025:.2f}')

sns.set_style('darkgrid')
axes = sns.regplot(x= louisville_tem.Date, y= louisville_tem.Temperature)
```
